```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
boston=load_boston()
df = pd.DataFrame(data=boston.data, columns = boston.feature_names)
df.head()
```

```python
df_scaled = pd.DataFrame(data=preprocessing.scale(boston.data), columns = boston.feature_names)
df_scaled.head()
```

```python
df.describe()
```

```python
df_scaled.describe()
```

```python
df_true_labels = pd.DataFrame(data=boston.target, columns=['target'])
```

```python
print(df_true_labels)
```

```python
cluster_model_2 = KMeans(n_clusters=2, init='k-means++')
cluster_labels_2 = cluster_model_2.fit_predict(df_scaled)

cluster_model_3 = KMeans(n_clusters=3, init='k-means++')
cluster_labels_3 = cluster_model_3.fit_predict(df_scaled)

cluster_model_4 = KMeans(n_clusters=4, init='k-means++')
cluster_labels_4 = cluster_model_4.fit_predict(df_scaled)

cluster_model_5 = KMeans(n_clusters=5, init='k-means++')
cluster_labels_5 = cluster_model_5.fit_predict(df_scaled)

cluster_model_6 = KMeans(n_clusters=6, init='k-means++')
cluster_labels_6 = cluster_model_6.fit_predict(df_scaled)
```

```python
silhouette_score_2 = metrics.silhouette_score(df_scaled, cluster_labels_2)
silhouette_score_3 = metrics.silhouette_score(df_scaled, cluster_labels_3)
silhouette_score_4 = metrics.silhouette_score(df_scaled, cluster_labels_4)
silhouette_score_5 = metrics.silhouette_score(df_scaled, cluster_labels_5)
silhouette_score_6 = metrics.silhouette_score(df_scaled, cluster_labels_6)
```

```python
print (silhouette_score_2)
print (silhouette_score_3)
print (silhouette_score_4)
print (silhouette_score_5)
print (silhouette_score_6)
```

```python
print(df_true_labels['target'])
```

```python
h_score2 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_2)
c_score2 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_2)

h_score3 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_3)
c_score3 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_3)

h_score4 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_4)
c_score4 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_4)

h_score5 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_5)
c_score5 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_5)

h_score6 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_6)
c_score6 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_6)

print (h_score2, c_score2)
print (h_score3, c_score3)
print (h_score4, c_score4)
print (h_score5, c_score5)
print (h_score6, c_score6)
```

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn import metrics

wine=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
wine.head()
```

```python
columns = ['Class','Alcohol', 'Malic_acid','Ash' ,'Alcalinity_of_ash','Magnesium','Total_phenols','Flavanoids', 'Nonflavanoid_phenols',
           'Proanthocyanins','Color_intensity','Hue','OD280_OD315_of_diluted_wines','Proline']
```

```python
wine.columns=columns
```

```python
wine.head()
```

```python
wine.describe()
```

```python
wine_scaled = pd.DataFrame(data=preprocessing.scale(wine.values), columns = wine.columns)
wine_scaled.head()
```

```python
wine_scaled.describe()
```

```python
df_true_labels = pd.DataFrame(wine, columns=['Class'])
```

```python
cluster_model = KMeans(n_clusters=3, init='k-means++')
cluster_labels = cluster_model.fit_predict(wine_scaled)
print(cluster_labels)
```

```python
silhouette_avg = metrics.silhouette_score(wine_scaled,cluster_labels)
print(silhouette_avg)
```

```python
target_labels = pd.qcut(x=df_true_labels,q=3,labels=[0,1,2,3], duplicates='drop' )
class_labels = df_true_labels.astype('int32') - 1
```

```python
h_score = metrics.homogeneity_score(df_true_labels,
                                    class_labels)
c_score = metrics.completeness_score(df_true_labels,
                                     class_labels)

print (h_score, c_score)
```

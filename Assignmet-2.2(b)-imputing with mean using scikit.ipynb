```python
import pandas as pd
import numpy as nm
from sklearn.preprocessing import Imputer

df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", delim_whitespace = True, header=None, 
               names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin', 'car_name'] )
print(df)
```

```python
df.horsepower=df.horsepower.replace(to_replace='?', value=nm.NaN)
df['horsepower'] = df['horsepower'].astype(float)
#df.describe()
print(df)
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=1, copy=False)
clean_data = imp.fit(df['horsepower'])
df['horsepower'].ix[374].dtype
clean_data
```
